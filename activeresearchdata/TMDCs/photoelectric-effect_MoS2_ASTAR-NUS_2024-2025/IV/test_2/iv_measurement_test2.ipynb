```python
# read csv and import package
import pandas as pd
import os

# Define the file name and folder name
file_name = 'extraction.xlsx'
output_folder = 'csv'

# Create the output folder
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Load the Excel file
data = pd.ExcelFile(file_name)

# Iterate through each sheet and save it as a separate CSV file
for sheet_name in data.sheet_names:
    sheet_data = data.parse(sheet_name)
    output_file = os.path.join(output_folder, f"{sheet_name}.csv")
    sheet_data.to_csv(output_file, index=False)

print(f"All sheets have been saved to the '{output_folder}' folder.")
```

```python
import os
import pandas as pd

# Define input and output folders
input_folder = 'reliable'
output_folder = 'reliable-csv'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate through all .xls files in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith('.xls'):
        # Construct full file path
        input_file_path = os.path.join(input_folder, file_name)
        
        # Read the Excel file
        data = pd.ExcelFile(input_file_path)
        
        # Iterate through each sheet in the Excel file
        for sheet_name in data.sheet_names:
            sheet_data = data.parse(sheet_name)
            
            # Construct output file path
            output_file_name = f"{os.path.splitext(file_name)[0]}_{sheet_name}.csv"
            output_file_path = os.path.join(output_folder, output_file_name)
            
            # Save the sheet as a CSV file
            sheet_data.to_csv(output_file_path, index=False)

print(f"All .xls files have been converted and saved to the '{output_folder}' folder.")
```
