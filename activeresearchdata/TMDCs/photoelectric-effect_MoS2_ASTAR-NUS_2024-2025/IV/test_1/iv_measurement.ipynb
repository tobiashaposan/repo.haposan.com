# I-V measurement breakdown
This Jupyter Notebook is to separate .xls file from each other.

Sample: MoS2

Measurement time: 13/2-25

```python
import pandas as pd
import os

# Define the file name and folder name
file_name = 'cntfet-vgs-id#1.xls'
output_folder = 'csv'

# Create the output folder
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Load the Excel file
data = pd.ExcelFile(file_name)

# Iterate through each sheet and save it as a separate CSV file
for sheet_name in data.sheet_names:
    sheet_data = data.parse(sheet_name)
    output_file = os.path.join(output_folder, f"{sheet_name}.csv")
    sheet_data.to_csv(output_file, index=False)

print(f"All sheets have been saved to the '{output_folder}' folder.")
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import os

# Define the folder name
csv_folder = 'csv'

# Plot all of the CSV files in the folder "csv"
# First column is x-axis titled "current (mA)", second column is y-axis titled "voltage (V)"
for file in os.listdir(csv_folder):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(csv_folder, file))
        plt.plot(df.iloc[:, 0], df.iloc[:, 1], label=file)

plt.xlabel("current (mA)")
plt.ylabel("voltage (V)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Plot each CSV file in the folder "csv" in a separate graph
# First column is x-axis titled "current (mA)", second column is y-axis titled "voltage (V)"
for file in os.listdir(csv_folder):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(csv_folder, file))
        plt.figure()
        plt.plot(df.iloc[:, 0], df.iloc[:, 1], color='green')  # Set data line color to green
        plt.xlabel("current (mA)")
        plt.ylabel("voltage (V)")
        plt.title(file)
        plt.grid(True)  # Add grid to the plot
        
        # Customize grid lines
        ax = plt.gca()
        ax.axhline(y=0, color='k', linewidth=2)  # Bold horizontal line at y=0
        ax.axvline(x=0, color='k', linewidth=2)  # Bold vertical line at x=0
        
        plt.show()
```
